# Visualizations

```python
def ptbarplot2(xdf, col, gvar, wvar, colors=['red', 'blue'], title='', sort_var='Percent_x', ascending=True, labels=None, ddi=None, n=10, output=False):
    if labels is None:
        labels = sorted(list(xdf[col].unique()))
        
    t0 = pt(ddi, xdf[xdf[col]==labels[0]], gvar, wvar)
    t1 = pt(ddi, xdf[xdf[col]==labels[1]], gvar, wvar)
    tm = t0.merge(t1, on=gvar).sort_values(by=sort_var, ascending=ascending)
    tm[['Percent_x', 'Percent_y']] = tm[['Percent_x', 'Percent_y']] * 100.
    try:
        ylabel = ddi.get_variable_info(gvar).label.title()
    except:
        ylabel = gvar
    
    g = tm.reset_index(drop=True).loc[0:n].sort_values(by=sort_var, ascending=True).plot.barh(
        x=gvar,
        y=['Percent_x','Percent_y'],
        color=colors,
        label=labels,
        title=title,
        figsize=(12, 8),
        legend='lower right',
    )

    for container in g.containers:
        g.bar_label(container, fmt='%.1f%%', padding=2)

    g.legend(labels, loc='lower right')
    g.set_ylabel(ylabel)
    g.set_xlabel("Within Group Percent of Data")
    if output:
        return tm
```

```python
xtm = ptbarplot2(
    xdf=df[df['Class_of_worker'] !='NIU'], 
    col='Hispanic or Not', 
    gvar='OCC2010', 
    wvar='ASECWT', 
    colors=['red', 'blue'],
    title='Class of Worker by Hispanic Response sorted by Top 25 Occupations of Hispanics', 
    ddi=ddi_codebook,
    ascending=False,
    n=25,
    output=True,
)
```

```python
def ptbarplot(ddi, df, xvar, wvar, color="blue", out=False):
    x = pt(ddi, df, xvar, wvar)
    try:
        ylabel = ddi.get_variable_info(xvar).label.title()
    except:
        ylabel = xvar
    x["Percent"] = (x["Percent"] * 100.0).round(2)

    y = (
        x[["Percent", xvar]]
        .sort_values(by="Percent", ascending=False)
        .reset_index(drop=True)
        .loc[0:30]
    )
```
